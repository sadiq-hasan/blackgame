```python
import random
class blackjack:
    total=10
    s = ['Diamonds', 'Hearts', 'Spades', 'Clubs']
    r= ['Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace']
    values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
    def __init__(self,a):
        self.a=a
        self.sum1=0
        self.sum2=0
        if self.a>self.total:
            print("invalid")
        else:
            print("best of luck")
            print("\n")
            
    def display(self):
        print("players 2 cards are")
        i=0
        while i<2:
            t=random.randint(0,3)
            m=random.randint(0,12)
            n=self.r[m]
            b=self.values[n]
            self.sum1+=b
            print(str(self.r[m])+" "+str(self.s[t]))
            i+=1
        print("current winning of the player is",self.sum1)
        print("\n")
        print("DEALERS 2 CARDS ARE")
        print("HIDDEN CARD")
        i=0
        while i<2:
            t=random.randint(0,3)
            m=random.randint(0,12)
            n=self.r[m]
            b=self.values[n]
            self.sum2+=b
            #print(str(self.r[m])+" "+str(self.s[t]))
            #print(i)
            if i==0:
                i+=1
                continue
            #print(sum2)
            #print("hey")
            #print(i)
            print(str(self.r[m])+" "+str(self.s[t]))
            i+=1
            
        print(self.sum2)
        
    def playeroption(self):
        d=1
        a=int(input("CHOOSE THE OPTION H OR S ? if h enter 1 else 0"))
        print(a)
        if a==1:
            self.onemore1()
        if a==0:
            self.stop()
            
    
    
    def onemore1(self):
        print("\n")
            
                
        t=random.randint(0,3)
        m=random.randint(0,12)
        n=self.r[m]
        b=self.values[n]
        self.sum1+=b
        print(str(self.r[m])+" "+str(self.s[t]))
        
        print("current winning of the player is",self.sum1)
    
        if self.sum1>21:
            print("player is busted")
        else:
            d=1
            a=int(input("CHOOSE THE OPTION H OR S ? if h enter 1 else 0"))
            print(a)
            if a==1:
                self.onemore1()
            if a==0:
                self.stop()
            
                
        
        
        
    def stop(self):
        print("NOW ITS DEALER OPTION")
        
        
            
    def dealeroption(self):
        d=1
        a=int(input("CHOOSE THE OPTION H OR S ? if h enter 1 else 0"))
        print(a)
        if a==1:
            self.onemore2()
        if a==0:
            self.stop2()
            
    
    
    def onemore2(self):
        print("\n")
            
                
        t=random.randint(0,3)
        m=random.randint(0,12)
        n=self.r[m]
        b=self.values[n]
        self.sum2+=b
        print(str(self.r[m])+" "+str(self.s[t]))
        
        print("current winning of the dealer is",self.sum2)
    
        if self.sum1>self.sum2:
            print("dealer is busted")
            self.done()
        elif self.sum1<self.sum2:
            print("player is busted")
        else:
            d=1
            a=int(input("CHOOSE THE OPTION H OR S ? if h enter 1 else 0"))
            print(a)
            if a==1:
                self.onemore1()
            if a==0:
                self.stop()
            
                
        
        
        
    def stop2(self):
        print("DONE")
        
        
    def done(self):
        print("GAME OVER")
        
        
        
    
    

    

```
